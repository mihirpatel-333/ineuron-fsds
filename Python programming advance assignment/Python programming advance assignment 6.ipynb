# 1. You are given two strings s and t. String t is generated by randomly <br>
shuffling string s and then adding one more letter at a random position. <br>
Return the letter that was added to t. <br>
Examples <br>
find_the_difference(&quot;abcd&quot;, &quot;abcde&quot;) ➞ &quot;e&quot; <br>
find_the_difference(&quot;&quot;, &quot;y&quot;) ➞ &quot;y&quot; <br>
find_the_difference(&quot;ae&quot;, &quot;aea&quot;) ➞ &quot;a&quot; <br>

```python
def find_the_difference(in_one,in_two):
    in_two_temp = list(in_two)
    for ele in in_one:
        if ele in in_two_temp:
            in_two_temp.pop(in_two_temp.index(ele))
    output = ','.join(in_two_temp)
    print(f'find_the_difference{in_one,in_two} ➞ {output}')

find_the_difference("abcd", "abcde")
find_the_difference("", "y")
find_the_difference("ae", "aea")
```

# 2. Given a function that accepts unlimited arguments, check and count how
many data types are in those arguments. Finally return the total in a list. <br>
List order is: <br>
[int, str, bool, list, tuple, dictionary] <br>
Examples <br>
count_datatypes(1, 45, &quot;Hi&quot;, False) ➞ [2, 1, 1, 0, 0, 0] <br>
count_datatypes([10, 20], (&quot;t&quot;, &quot;Ok&quot;), 2, 3, 1) ➞ [3, 0, 0, 1, 1, 0] <br>
count_datatypes(&quot;Hello&quot;, &quot;Bye&quot;, True, True, False, {&quot;1&quot;: &quot;One&quot;, &quot;2&quot;: &quot;Two&quot;}, [1, <br>
3], {&quot;Brayan&quot;: 18}, 25, 23) ➞ [2, 2, 3, 1, 0, 2] <br>
count_datatypes(4, 21, (&quot;ES&quot;, &quot;EN&quot;), (&quot;a&quot;, &quot;b&quot;), False, [1, 2, 3], [4, 5, 6]) ➞ [2, <br>
0, 1, 2, 2, 0]

```python
from collections import OrderedDict
def count_datatypes(*args):
    output = OrderedDict({'int':0,'str':0,'bool':0,'list':0,'tuple':0,'dict':0})
    for ele in args:
        output[type(ele).__name__] += 1
    print(f'count_datatypes{args}  ➞  {list(output.values())}')
    
count_datatypes(1, 45, "Hi", False)   
count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1)
count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23)
count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6])
```

# 4. Given an integer between 0 and 26, make a variable (self.answer). That
variable would be assigned to a string in this format: <br>
&quot;nines:your answer, threes:your answer, ones:your answer&quot; <br>
You need to find out how many ones, threes, and nines it would at least take <br>
for the number of each to add up to the given integer when multiplied by one, <br>
three or nine (depends). <br>
Examples <br>
ones_threes_nines(10) ➞ &quot;nines:1, threes:0, ones:1&quot; <br>
ones_threes_nines(15) ➞ &quot;nines:1, threes:2, ones:0&quot; <br>
ones_threes_nines(22) ➞ &quot;nines:2, threes:1, ones:1&quot; <br>

```python
def ones_threes_nines(in_num):
    in_num_clone = in_num
    answer = {'nines':0,'threes':0,'ones':0}
    if in_num > 0 and in_num < 26:
        while in_num != 0:
            if in_num >= 9:
                in_num -= 9
                answer['nines'] += 1
            elif in_num >= 3:
                in_num -= 3
                answer['threes'] += 1
            elif in_num >= 1:
                in_num -=1
                answer['ones'] +=1
    print(f"ones_threes_nines({in_num_clone}) ➞ \"nines:{answer['nines']}, threes:{answer['threes']}, ones:{answer['ones']}\"")
    
ones_threes_nines(10) 
ones_threes_nines(15)
ones_threes_nines(22)
```

# 5. The Fibonacci sequence is a classic use case for recursive functions since
the value of the sequence at a given index is dependent on the sum of the <br>
last two values. However, the recursion tree created by solving the Fibonacci <br>
sequence recursively can grow quite fast. Therefore it can be important to <br>
think about the implications of running a function recursively. Depending on <br>
the size of n needed and the capabilities of the system in question you might <br>
want to take a different approach. <br>
Write a non-recursive function that takes an integer n and returns the <br>
Fibonacci sequence&#39;s value at index n. <br>
Examples <br>
fib(6) ➞ 8 <br>
# 0 + 1 = 1, 1 + 1 = 2, 1 + 2 = 3, 2 + 3 = 5, 3 + 5 = 8 <br>
fib(1) ➞ 1 <br> 
fib(2) ➞ 1

```python
# Approach 1
def fib_one(in_num):
    if in_num < 1:
        print("Minimum two digits are required to generate a fibonacci sequence")
    else:
        output = [0,1]
        for ele in range(2,in_num+1):
            output.append(output[-1]+output[-2])
        print(f'fib_one({in_num}) ➞ {output[-1]}')

# Approach 2
def fib_two(in_num):
    back_one,back_two = 1,0
    for ele in range(2,in_num+1):
        temp = back_one+back_two
        back_two = back_one
        back_one = temp
    print(f'fib_two({in_num}) ➞ {back_one}')

fib_one(6)
fib_one(1)
fib_one(2)
print()
fib_two(6)
fib_two(1)
fib_two(2)
```
